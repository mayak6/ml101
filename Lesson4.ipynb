```
caps = cv.imread('caps.jpg')
# res = cv.resize(caps,None,fx=2, fy=2, interpolation = cv.INTER_CUBIC)

caps_scaled = cv.resize(caps,(600, 800), interpolation = cv.INTER_CUBIC)

hsv = cv.cvtColor(caps, cv.COLOR_BGR2HSV)

lower_red = np.array([204,0,0])
upper_red = np.array([255,105,97])

mask = cv.inRange(hsv, lower_red, upper_red)

red_cap = cv.bitwise_and(caps, caps, mask = mask)

cv2_imshow(caps_scaled)
cv2_imshow(mask)
cv2_imshow(red_cap)
```
